```python
import cv2
```

```python
image = cv2.imread('/home/shivanisri/Desktop/Shivani/vallourec_images/2022-08-02-06-05-50-000010423-cam1.png')
template = cv2.imread('/home/shivanisri/Desktop/Shivani/template.png')
print(template.shape)
print(image.shape)
```

```python
up_width = 640
up_height = 480
up_points = (up_width, up_height)
resized_up = cv2.resize(template, up_points, interpolation= cv2.INTER_LINEAR)
print(resized_up.shape)
#cv2.imshow('resized',resized_up)
```

```python
subtracted = cv2.subtract(image,resized_up)
```

```python
cv2.imshow('image', subtracted)
```
